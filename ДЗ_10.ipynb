```python
power = pd.read_csv('power.csv')
baltic = power[ (power['country']=='Latvia') 
               | (power['country']=='Lithuania') 
               | (power['country']=='Estonia') ]
baltic_filter = baltic[ (baltic['quantity'] > 0)
       & ((baltic['year'] >= 2005)
          & (baltic['year'] <= 2010))
       & ((baltic['category'] == 4) | (baltic['category'] == 12) | (baltic['category'] == 21))]

print(f"Суммарное энергопотребление Прибалтики: {baltic_filter['quantity'].sum()}")


```
